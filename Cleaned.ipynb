# **Stage 3: Chia dá»¯ liá»‡u (Train / Test Split)**

**`PhÆ°Æ¡ng Ã¡n: Stratified Split (PhÃ¢n chia theo táº§ng)`**

`ÄÃ¢y lÃ  cÃ¡ch há»£p lÃ½ nháº¥t cho bÃ i toÃ¡n há»“i quy giÃ¡ nhÃ  nÃ y náº¿u má»¥c tiÃªu cá»§a báº¡n lÃ  xÃ¢y dá»±ng mÃ´ hÃ¬nh Ä‘á»‹nh giÃ¡ chung.`

* **Táº¡i sao khÃ´ng nÃªn chia ngáº«u nhiÃªn (Random Split)?** Náº¿u chia ngáº«u nhiÃªn, cÃ³ kháº£ nÄƒng táº­p Test sáº½ khÃ´ng chá»©a cÄƒn nhÃ  "siÃªu sang" nÃ o, hoáº·c chá»©a toÃ n bá»™ cÃ¡c cÄƒn nhÃ  giÃ¡ ráº». Khi Ä‘Ã³, mÃ´ hÃ¬nh sáº½ Ä‘Ã¡nh giÃ¡ sai lá»‡ch Ä‘á»™ chÃ­nh xÃ¡c thá»±c táº¿.

* **CÃ¡ch thá»±c hiá»‡n**: ChÃºng ta sáº½ chia dá»¯ liá»‡u sao cho tá»· lá»‡ phÃ¢n bá»‘ giÃ¡ nhÃ  á»Ÿ táº­p Train vÃ  táº­p Test lÃ  tÆ°Æ¡ng Ä‘á»“ng nhau. VÃ­ dá»¥: Náº¿u táº­p dá»¯ liá»‡u gá»‘c cÃ³ 10% nhÃ  giÃ¡ cao, thÃ¬ táº­p Train vÃ  táº­p Test cÅ©ng pháº£i Ä‘áº£m báº£o cÃ³ khoáº£ng 10% nhÃ  giÃ¡ cao.

```python
from sklearn.model_selection import StratifiedShuffleSplit

# 2. Táº¡o má»™t cá»™t phÃ¢n nhÃ³m (category) cho SalePrice Ä‘á»ƒ phÃ¢n táº§ng
# Chia SalePrice thÃ nh 5 nhÃ³m (bins) Ä‘á»ƒ Ä‘áº£m báº£o má»—i nhÃ³m cÃ³ Ä‘á»§ Ä‘áº¡i diá»‡n
# Labels=1 lÃ  giÃ¡ tháº¥p nháº¥t, Labels=5 lÃ  giÃ¡ cao nháº¥t
df["price_cat"] = pd.qcut(df["SalePrice"], q=5, labels=[1, 2, 3, 4, 5])

# 3. Sá»­ dá»¥ng StratifiedShuffleSplit
# Chia 80% Train - 20% Test
split = StratifiedShuffleSplit(n_splits=1, test_size=0.2, random_state=42)

for train_index, test_index in split.split(df, df["price_cat"]):
    train_set = df.loc[train_index]
    test_set = df.loc[test_index]

# 4. TÃ¡ch X (Features) vÃ  y (Target)
# LÆ°u Ã½: Drop cáº£ 'price_cat' vÃ¬ nÃ³ chá»‰ lÃ  biáº¿n phá»¥ trá»£
X_train = train_set.drop(["SalePrice", "price_cat"], axis=1)
y_train = train_set["SalePrice"].copy()

X_test = test_set.drop(["SalePrice", "price_cat"], axis=1)
y_test = test_set["SalePrice"].copy()

# 5. Kiá»ƒm tra kÃ­ch thÆ°á»›c (Optional)
print("Shape of X_train:", X_train.shape)
print("Shape of y_train:", y_train.shape)
print("Shape of X_test:", X_test.shape)
print("Shape of y_test:", y_test.shape)
print(train_set.describe())
print(test_set.describe())
```

**`=> Táº­p train vÃ  test cÃ³ thá»‘ng kÃª mÃ´ táº£ gáº§n nhÆ° lÃ  giá»‘ng nhau, chá»©ng tá» chiáº¿n thuáº­t chia Ä‘Ã£ phÃ¡t huy tÃ¡c dá»¥ng`**

## **4.4 Correlation Checking**

`Äá»‘i vá»›i cÃ¡c má»‘i quan há»‡ tÆ°Æ¡ng quan Ä‘á»‘i vá»›i biáº¿n má»¥c tiÃªu, má»‘i quan há»‡ nÃ o cÃ³ tÆ°Æ¡ng quan cÃ ng cao Ä‘á»“ng nghÄ©a vá»›i kháº£ nÄƒng dá»± bÃ¡o cÃ ng chÃ­nh xÃ¡c.
```python
target_var = 'SalePrice'
corr_with_target = corr_matrix[target_var].sort_values(ascending=False)
corr_with_target = corr_with_target.drop(target_var)

plt.figure(figsize=(12, 10))
ax = sns.barplot(x=corr_with_target.values, y=corr_with_target.index, palette='vlag')
plt.title(f'Má»©c Ä‘á»™ TÆ°Æ¡ng quan cá»§a cÃ¡c Biáº¿n vá»›i "{target_var}"', fontsize=16)
plt.xlabel('Há»‡ sá»‘ TÆ°Æ¡ng quan Pearson')
plt.ylabel('Biáº¿n')
plt.grid(axis='x', linestyle='--', alpha=0.6)
for p in ax.patches:
    width = p.get_width()
    y = p.get_y() + p.get_height() / 2
    if width > 0:
        x = width + 0.01
        ha = 'left'
    else:
        x = width - 0.01
        ha = 'right'
    ax.text(
        x=x,                 # Vá»‹ trÃ­ x
        y=y,                 # Vá»‹ trÃ­ y
        s=f'{width:.2f}',    # VÄƒn báº£n cáº§n hiá»ƒn thá»‹ (lÃ m trÃ²n 2 chá»¯ sá»‘)
        ha=ha,               # CÄƒn lá» ngang
        va='center',         # CÄƒn lá» dá»c
        fontsize=10
    )
plt.show()
```

```python
threshold = 0.05

high_corr_features = corr_with_target[abs(corr_with_target) < threshold]

sorted_corr = high_corr_features.sort_values(ascending=False)

print(f"--- CÃC BIáº¾N CÃ“ TÆ¯Æ NG QUAN TUYá»†T Äá»I Vá»šI '{target_var}' < {threshold} ---")
print("(Dáº¥u hiá»‡u cá»§a cÃ¡c biáº¿n cÃ³ tÆ°Æ¡ng quan yáº¿u vá»›i biáº¿n má»¥c tiÃªu)")
if sorted_corr.empty:
    print("\nKhÃ´ng tÃ¬m tháº¥y biáº¿n nÃ o vÆ°á»£t ngÆ°á»¡ng.")
else:
    print(sorted_corr.round(2))
```

```python
criteria = {
    'Gr Liv Area': 4000,   # NhÃ  quÃ¡ to > 4000 sqft
    'Lot Area': 60000,    # Äáº¥t quÃ¡ rá»™ng > 60000 sqft
    'Total Bsmt SF': 3000, # Háº§m quÃ¡ lá»›n > 3000 sqft
    '1st Flr SF': 4000     # Táº§ng 1 quÃ¡ rá»™ng > 4000 sqft
}

print("--- BÃO CÃO PHÃT HIá»†N OUTLIER ---")
fig, axes = plt.subplots(2, 2, figsize=(14, 10))
axes = axes.flatten()

for i, (feature, thresh) in enumerate(criteria.items()):
    # Náº¿u cá»™t Ä‘Ã£ bá»‹ loáº¡i bá» trÆ°á»›c Ä‘Ã³ thÃ¬ bá» qua vÃ  Ä‘Ã¡nh dáº¥u Ã´ tÆ°Æ¡ng á»©ng
    if feature not in X_train.columns:
        print(f"- {feature}: Cá»™t khÃ´ng tá»“n táº¡i trong X_train (cÃ³ thá»ƒ Ä‘Ã£ bá»‹ drop). Bá» qua.")
        axes[i].axis('off')
        axes[i].set_title(f'{feature} - DROPPED')
        continue

    # Äáº¿m sá»‘ lÆ°á»£ng vi pháº¡m
    outlier_count = len(X_train[X_train[feature] > thresh])
    print(f"- {feature}: PhÃ¡t hiá»‡n {outlier_count} trÆ°á»ng há»£p > {thresh}")
    
    # Váº½ biá»ƒu Ä‘á»“
    sns.scatterplot(x=X_train[feature], y=y_train, ax=axes[i], alpha=0.6, color='steelblue')
    # Váº½ Ä‘Æ°á»ng ranh giá»›i Ä‘á»
    axes[i].axvline(x=thresh, color='red', linestyle='--', linewidth=2, label=f'NgÆ°á»¡ng cáº¯t ({thresh})')
    
    # TÃ´ mÃ u vÃ¹ng bá»‹ loáº¡i bá» Ä‘á»ƒ dá»… nhÃ¬n
    axes[i].axvspan(thresh, X_train[feature].max(), color='red', alpha=0.1)
    
    axes[i].set_title(f'{feature} vs SalePrice')
    axes[i].legend()

plt.tight_layout()
plt.show()
```

# **Stage 4: Data Preprocessing Pipeline**

```python
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer

numeric_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='median')), # Äiá»n missing báº±ng trung vá»‹
    ('scaler', StandardScaler())                   # Chuáº©n hÃ³a
])

categorical_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='constant', fill_value='missing')), # Äiá»n missing
    ('encoder', OneHotEncoder(handle_unknown='ignore', sparse_output=False)) # MÃ£ hÃ³a One-hot
])

preprocessor = ColumnTransformer(
    transformers=[
        ('num', numeric_transformer, numeric_cols_of_X_train_clean),
        ('cat', categorical_transformer, categorical_cols_of_X_train_clean)
    ])
```

# **Stage 5 & 6: Feature Selection & Modeling**

```python
from sklearn.linear_model import LinearRegression
from sklearn.feature_selection import SelectPercentile, f_regression
from sklearn.model_selection import cross_val_score
from sklearn.metrics import mean_squared_error, r2_score
import matplotlib.pyplot as plt

model_pipeline = Pipeline(steps=[
    ('preprocessor', preprocessor),
    ('feature_selection', SelectPercentile(f_regression, percentile=50)), # Giá»¯ láº¡i 50% Ä‘áº·c trÆ°ng tá»‘t nháº¥t
    ('regressor', LinearRegression())
])

cv_scores = cross_val_score(model_pipeline, X_train_clean, y_train_clean, cv=5, scoring='neg_mean_squared_error')
rmse_cv = np.sqrt(-cv_scores)

print(f"Cross-Validation RMSE: {rmse_cv.mean():.2f}")
```

# **Stage 7: Final Evaluation**

`Kiá»ƒm tra nÄƒng lá»±c thá»±c táº¿ trÃªn Test Set (Dá»¯ liá»‡u chÆ°a bao giá» Ä‘Æ°á»£c nhÃ¬n tháº¥y).`

```python
from sklearn.metrics import mean_absolute_error
# 1. Huáº¥n luyá»‡n láº¡i trÃªn toÃ n bá»™ táº­p Train
model_pipeline.fit(X_train_clean, y_train_clean)

# 2. Dá»± bÃ¡o trÃªn táº­p Test
y_pred_lr = model_pipeline.predict(X_test)

#y_pred_real = np.expm1(y_pred_lr) 
y_test_log = np.log1p(y_test)
# Chuyá»ƒn Ä‘á»•i ngÆ°á»£c láº¡i tá»« log
# 3. ÄÃ¡nh giÃ¡ chá»‰ sá»‘
final_rmse_lr = np.sqrt(mean_squared_error(y_test_log, y_pred_lr))
final_r2_lr = r2_score(y_test_log, y_pred_lr)
mae = mean_absolute_error(y_test_log, y_pred_lr)

print("------------------------------------------------")
print(f"Linear Regression - Test RMSE: {final_rmse_lr:.2f}")
print(f"Linear Regression - R2 Score: {final_r2_lr:.2f}")
print(f"Linear Regression - MAE: {mae:.2f}")
print("------------------------------------------------")
```
